# Self Study Case Study - Enron Network Analysis

In this self study case study, we will try to understand the employee network in Enron on a sample of Enron dataset.

## Background

The story of Enron is a story of a company that reached from immense heights and deepest lows in no time.

Enron Corp. was one of the biggest firms in the United States and was delivering splendid performance on the wall street. However, the company witnesses sudden drop in the prices and declared bankruptcy. How one of the most powerful business in the US and the world, disintegrated overnight is still a puzzle to many. The Enron leadership was involved in one of the biggest frauds and this particular fraud has been an area of interest for many researchers and the ML practioners.

In this case study, we have a subset for 50 senior officials and the idea is to build a network from the mails that were sent and received to better understand the connections and highlight the important nodes (people) in this group.

## Steps

- Read the data and understand the structure of data
- Put the data into a graph
- Identify the important nodes from the visualization
- Calculate the centrality measures and quantify the importance 
- Highlight the important nodes through color coding and comment on the roles/importance that can be figured out from this.

### Import the files

```
import pandas as pd

data = pd.read_csv('/content/drive/My Drive/EmailEnron.csv')
```

```
data.shape
```

```
import networkx as nx

G = nx.Graph()
```

```
G = nx.from_pandas_edgelist(data, 'From', 'To')

import matplotlib.pyplot as plt
plt.figure(figsize = (10,10))

options = {
    "node_color": "black",
    "node_size": 10,
    "linewidths": 0.5,
    "width": 0.1,
}
nx.draw(G, with_labels = True, **options)
plt.show()
```

```
plt.figure(figsize = (10,10))
nx.draw_shell(G, with_labels = True, **options)
```

```
plt.figure(figsize = (10,10))
nx.draw_spring(G, with_labels = True) # with the default parameters
```

### Comments from the Visualization

- Of the 80 nodes in the dataset, 1 appears to be the most important node and is connected with all the other nodes. We can interpret this official, perhaps as the CEO.
- Other nodes that are important are also highlighted in the visualisation - 56,54,74,53,50. The circular visualisation is a better visualisation approach to cleary highlight the important nodes.
- There are internal team structures that appear from the visualisation but are not very clear as to what nodes are part of the teams.

```
# Let us quickly look at the degree of the nodes
for i in G.degree():
  print(i)
```

### Centrality measures

```
deg_cen = nx.degree_centrality(G)
eig_cen = nx.eigenvector_centrality(G)
clo_cen = nx.closeness_centrality(G)
bet_cen = nx.betweenness_centrality(G)
```

#### Degree Centrality

```
temp = {}

for w in sorted(deg_cen, key=deg_cen.get, reverse=True):
    temp[w] = deg_cen[w]
print("Sorted Importance of nodes in terms of deg_cen for Phase {} is {}".format(w+1, list(temp.keys())[:5]))
print()
```

```
# let us now, color these nodes and visualise the network again

color = []
for node in G:
    if (node == 1 or node == 56  or node == 74 or node==53 or node==54):
        color.append('red')
    else:
        color.append('black')
plt.figure(figsize=(10,10))
nx.draw(G,node_color=color,with_labels=True)
```

#### Eigen Vector Centrality

```
temp = {}

for w in sorted(eig_cen, key=eig_cen.get, reverse=True):
    temp[w] = eig_cen[w]
print("Sorted Importance of nodes in terms of eig_cen for Phase {} is {}".format(w+1, list(temp.keys())[:5]))
print()
```

```
# let us now, color these nodes and visualise the network again

color = []
for node in G:
    if (node == 1 or node == 56  or node == 74 or node==53 or node==54):
        color.append('red')
    else:
        color.append('black')
plt.figure(figsize=(10,10))
nx.draw(G,node_color=color,with_labels=True)
```

#### Betweenness Centrality


```
temp = {}

for w in sorted(bet_cen, key=bet_cen.get, reverse=True):
    temp[w] = bet_cen[w]
print("Sorted Importance of nodes in terms of bet_cen is {}".format(list(temp.keys())[:5]))
print()
```
